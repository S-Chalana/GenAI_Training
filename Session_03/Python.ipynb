## Can you write a program that generates this pyramid with a N value of 10? (X'mas tree)

```python
## left aligned triangle
def art_design(num):
    for i in range(1, num):
        print(" "+"*"*i )

art_design(10)
```

```python
## right aligned triangle
def art_design_right(num):
    for i in range(1, num):
        print(" " * (num - i) + "*" * i )

art_design_right(10)
```

```python
## pyramid
def art_design_pyramid(num):
    for i in range(1, num):
        print(" " * (num - i) +"*"*(2*i-1) )

art_design_pyramid(10)
```
