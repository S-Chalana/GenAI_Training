```python
### Build a number guessing game that keeps looping until the user guesses correctly.

lucky_number = 5

def guess():
    num = int(input("Enter any number (between 1 to 9)"))
    return num

for i in range(1,9):
    num = guess()
    if num == lucky_number:
        print("you guessed the right number")
        break
    elif num < 1 or num > 9:
        print("Number should be between 1 and 9")
    elif num >=1 and num <=9:
        print("Guess again")
      
       
    
    
```

```python
given_list = {1, 2, 3, 4, 6, 7, 8, 9, 10}

def find_missing_num(num_list):
    missing_num = 0
    
    for i in num_list:
            missing_num = missing_num+1
            if (missing_num != i):
                print(f"The missing number is {missing_num}")
                break
            else:
                print(missing_num, i)
    
        
find_missing_num(given_list)

        
```

```python
from collections import Counter

given_str = "AI makes AI easier and also AI takes jobs away! You got to learn AI to beat AI"

def word_count(text):
    words = text.split()           # Split into words
    counts = Counter(words)        # Count occurrences efficiently
    for i, count in counts.items():
        print(f"{i} = {count}")


word_count(given_str)
    
```
